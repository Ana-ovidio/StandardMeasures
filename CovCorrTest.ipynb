```python
import numpy as np
import pandas as pd
from tqdm import tqdm
```

```python
stocks_frame = pd.DataFrame(np.random.randn(366,100))
display(stocks_frame)
```

-----------
**Estima a variação entre os dias**

```python
variations = []
for i in range(1,len(stocks_frame)):
    variation = stocks_frame.iloc[i] - stocks_frame.iloc[i-1]
    variation_dic = variation.to_dict()
    variations.append(variation_dic) 
```

```python
stcks_prcs_vr = pd.DataFrame(variations)
```

```python
stcks_prcs_vr
```

-------------
**Estima a matriz de covariância e a covariância média**

```python
list_prices = []
for i in range(len(stcks_prcs_vr.columns)):
    list_prices.append(stcks_prcs_vr.iloc[:][i].tolist())
matrix_prices = np.array(list_prices)
```

```python
cov_matrix = np.cov(matrix_prices,bias=True)
corr_matrix = np.corrcoef(matrix_prices)
```

```python
valeus_outside_diagonal = []
for i, dataset in enumerate(cov_matrix):
    for j in range(i):
        valeus_outside_diagonal.append(dataset[j])
```

```python
mean_cov = np.mean(valeus_outside_diagonal)
mean_cov
```

```python
all_cov_mean = []
days = 9000
```

```python
for day in tqdm(range(days)):
    stocks_frame = pd.DataFrame(np.random.randn(366,100))
    
    variations = []
    for i in range(1,len(stocks_frame)):
        variation = stocks_frame.iloc[i] - stocks_frame.iloc[i-1]
        variation_dic = variation.to_dict()
        variations.append(variation_dic) 
        
    stcks_prcs_vr = pd.DataFrame(variations) 
    
    list_prices = []
    for i in range(len(stcks_prcs_vr.columns)):
        list_prices.append(stcks_prcs_vr.iloc[:][i].tolist())
    matrix_prices = np.array(list_prices)
    cov_matrix = np.cov(matrix_prices,bias=True)
    
    valeus_outside_diagonal = []
    for i, dataset in enumerate(cov_matrix):
        for j in range(i):
            valeus_outside_diagonal.append(dataset[j])
            
    mean_cov = np.mean(valeus_outside_diagonal)
    all_cov_mean.append(mean_cov)
```
